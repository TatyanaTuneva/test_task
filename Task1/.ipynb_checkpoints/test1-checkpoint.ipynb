```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Считываем файл

```python
data = pd.read_excel('Данных для тестового задания на должность аналитика #1.xls', header=1)
```

Нулевая гипотеза: корреляция между посещениями раздела и всего проекта в целом близка к 0 \
Альтернативная гипотеза: существует довольно большая корреляция между посещениями раздела и всего проекта в целом

Проверяем нулевую гипотезу

Создаем dataset с данными о посещаемости по неделям

```python
frame = {'Дата': data['Дата'], 'Посетители': data['Посетители'], 'Посетители раздела': data['Посетители.1']}
data_weeks = pd.DataFrame(frame)
```

Создаем dataset с данными о посещаемости по пиковым дням

```python
frame = {'Дата': data['Unnamed: 3'], 'Посетители': data['Unnamed: 4'], 'Посетители раздела': data['Unnamed: 11']}      
data_days = pd.DataFrame(frame)
```

График посещаемости по неделям

```python
ax = sns.lineplot(data=data_weeks, x='Дата', y='Посетители')
ax = sns.lineplot(data=data_weeks, x='Дата', y='Посетители раздела')
ax.set(title='График посещаемости по неделям')
```

График посещаемости по пиковым дням

```python
ax = sns.lineplot(data=data_days, x='Дата', y='Посетители')
ax = sns.lineplot(data=data_days, x='Дата', y='Посетители раздела')
ax.set(title='График посещаемости по пиковым дням')
```

По графику выше заметна корреляция между посещениями проекта и раздела, строим график, чтобы посмотреть взаимосвязь

```python
ax = sns.lmplot(x='Посетители раздела', y='Посетители', data=data_days)
ax.set(title='График взаимосвязи посетителей раздела и посетителей проекта по пиковым дням')
```

Находим коэффициент корреляции

```python
corr = round(data_days['Посетители раздела'].corr(data_days['Посетители']), 4) 
corr
```

Коэффициент 0.992 количественно выражает сильную, положительную корреляцию, уже наблюдавшуюся нами на графике \
Отклоняем нулевую гипотезу, принимаем альтернативную \
Наблюдается положительная взаимосвязь между количеством посетителей раздела и посетителей проекта в целом \
На графике 'График посещаемости по неделям' 2017-11-21 заметен пик посещаемости проекта в целом, следовательно посещаемость раздела тоже выросла
