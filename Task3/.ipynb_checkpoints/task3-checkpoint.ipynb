Выгружаем данные в dataset

```python
query = '''SELECT 
               reviews_marks.id, 
               avg_mark, 
               postdate,
               newfirms.name
           FROM 
               reviews_marks
           LEFT JOIN newfirms
               ON firmid = newfirms.id
           WHERE
              postdate BETWEEN "2019-04-01 00:00:00" AND "2019-06-30 00:00:00" AND
              is_deleted = 0
           ORDER BY 
               postdate;'''

SQL_Query = pd.read_sql_query(
    query,
    dbcon
)

data = pd.DataFrame(SQL_Query, columns=['id', 'avg_mark', 'postdate', 'name'])
```

Преобразуем даты

```python
data['postdate'] = pd.to_datetime(data['postdate']).apply(lambda x: '{year}-{month}'.format(year=x.year, month=x.month))
```

Находим медиану для каждой марки в каждый месяц

```python
data_median = data.groupby(['postdate','name'], as_index=False) \
                  .aggregate({'avg_mark': 'median'}) \
                  .rename(columns={'avg_mark': 'median'})
```

Находим количество оценок для каждой марки в каждый месяц

```python
data_count = data.groupby(['postdate','name'], as_index=False) \
                  .aggregate({'avg_mark': 'count'}) \
                  .rename(columns={'avg_mark': 'count'})
```

Соединяем data_median и data_count

```python
data_median_count = pd.merge(data_count, data_median, how='inner', on=['name', 'postdate']) 
```

Нулевая гипотеза: корреляция между количеством оценок и их медианой близка к 0 \
Альтернативная гипотеза: существует довольно большая корреляция между количеством оценок и их медианой

Проверяем нулевую гипотезу

Находим записи с максимальным количеством оценок, они сильнее всего будут влиять на общую медиану

```python
data_names_count_of_marks_max = data.groupby(['postdate', 'name'], as_index=False) \
                          .aggregate({'id': 'count'}) \
                          .groupby(['name'], as_index=False) \
                          .aggregate({'id': 'sum'}) \
                          .sort_values(['id']) \
                          .tail(8)['name']

data_names_count_of_marks_max_dataset = pd.merge(data_names_count_of_marks_max, data_median_count, how='inner', on=['name']) 
```

Строим график взаимосвязи медианы и количества оценок для каждой марки с максимальным количеством оценок в каждый месяц

```python
sns.set(rc={'figure.figsize':(40.,10.)})
ax = sns.lmplot(x='count', y='median', hue='name', data=data_names_count_of_marks_max_dataset)
ax.set_axis_labels(x_var='Количество оценок', y_var='Медиана оценок')
```

Проверяем корреляцию между количеством оценок и медианой оценок 

```python
corr = round(data_median_count['count'].corr(data_median_count['median']), 4) 
corr
```

Корреляция минимальна, значит количество оценок не влияет на медиану, принимаем нулевую гипотезу

Пробуем проверить изменение медианы оценок по каждому производителю

```python
sns.set(rc={'figure.figsize':(100.,50.)})
ax = sns.histplot(x='name', y='median', data=data_median_count.loc[data_median_count['postdate'] == '2019-4'],  color='red')
ax = sns.histplot(x='name', y='median', data=data_median_count.loc[data_median_count['postdate'] == '2019-6'])
ax.set_xlabel('Производитель')
ax.set_ylabel('Медиана оценки')
```

По графику видно, что в среднем оценки по каждой марке уменьшились, так же мы наблюдаем, что некоторые автомобили в конце периода только получили свои оценки, поэтому общая медиана оценок упала
